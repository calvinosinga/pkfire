```python
from pkfire.grid_lib.grid import Grid
import pkfire.dist_lib.dist as dlib
from pkfire.pkfire import Pkfire
from pkfire.vis_lib.plot import PKFPlot

import matplotlib.pyplot as plt
```

```python
grid = Grid((100, 100), 75)
dlib.gaussian(grid)
pkf = Pkfire(grid)
```

```python
pkf.pk()
```

```python
pkplt = PKFPlot(pkf)
fig, axes = plt.subplots(1, 3, gridspec_kw={'width_ratios':[1, 1, 0.15], 'wspace':0.35}, figsize = (7, 3))
pkplt.gridPlot(axes[1], axes[2])
pkplt.pkPlot(axes[0])
```
